### 4. Número de títulos agregados por año

```python
# Convert 'date_added' to datetime objects, handling mixed formats
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed', errors='coerce')

# Drop rows where 'date_added' could not be parsed (NaT values)
df.dropna(subset=['date_added'], inplace=True)

# Extract the year from 'date_added'
df['year_added'] = df['date_added'].dt.year

# Count the number of titles added each year and sort by year
titles_added_per_year = df['year_added'].value_counts().sort_index()

# Plotting the time series
plt.figure(figsize=(12, 6))
sns.lineplot(x=titles_added_per_year.index, y=titles_added_per_year.values, marker='o')
plt.title('Número de Títulos Agregados a Netflix por Año')
plt.xlabel('Año')
plt.ylabel('Cantidad de Títulos Agregados')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("\nAnálisis del crecimiento del catálogo:\n" \
      "El gráfico muestra un crecimiento constante en el número de títulos agregados anualmente, \n" \
      "con un pico significativo alrededor de los años 2018-2020. \n" \
      "Esto podría estar relacionado con la expansión global de Netflix y su inversión en contenido original en ese período, \n" \
      "así como la competencia creciente en el mercado de streaming.")
```

```python
# Ensure 'date_added' is in datetime format (already done in previous step, but good for robustness)
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed', errors='coerce')
df.dropna(subset=['date_added'], inplace=True)

# Extract month and year from 'date_added'
df['month_added'] = df['date_added'].dt.month_name()
df['year_added'] = df['date_added'].dt.year

# Order months for better visualization
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

# Create a pivot table for the heatmap
heatmap_data = df.groupby(['year_added', 'month_added']).size().unstack(fill_value=0)
heatmap_data = heatmap_data[month_order] # Ensure month order

# Plotting the heatmap
plt.figure(figsize=(15, 8))
sns.heatmap(heatmap_data, cmap='viridis', fmt='d', annot=True, linewidths=.5, cbar_kws={'label': 'Cantidad de Títulos Agregados'})
plt.title('Mapa de Calor de Títulos Agregados a Netflix por Año y Mes')
plt.xlabel('Mes')
plt.ylabel('Año')
plt.yticks(rotation=0)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\nAnálisis del Mapa de Calor de Lanzamientos:\n" \
      "El mapa de calor revela patrones de estacionalidad en la adición de contenido a Netflix. \n" \
      "A simple vista, se puede observar que ciertos meses, como diciembre y enero, \n" \
      "así como julio y octubre, tienden a tener un mayor número de estrenos. \n" \
      "Esto podría indicar estrategias de lanzamiento para captar audiencias durante periodos vacacionales \n" \
      "o de mayor consumo de entretenimiento en casa. Por otro lado, algunos meses muestran consistentemente \n" \
      "un menor número de adiciones. Las conclusiones sobre la estacionalidad sugieren que Netflix \n" \
      "optimiza sus lanzamientos en función de la demanda y el comportamiento del consumidor a lo largo del año.")
```

```python
# Filter for movies only
movies_df = df[df['type'] == 'Movie'].copy()

# Extract duration in minutes and convert to numeric
# Handle potential errors during conversion by coercing to NaN and then dropping them
movies_df['duration_minutes'] = pd.to_numeric(movies_df['duration'].str.replace(' min', ''), errors='coerce')
movies_df.dropna(subset=['duration_minutes'], inplace=True)

# Extract the main genre (the first one listed)
movies_df['main_genre'] = movies_df['listed_in'].apply(lambda x: x.split(',')[0].strip() if pd.notna(x) else 'Unknown')

# Select top N genres for better visualization, otherwise the plot might be too crowded
top_genres = movies_df['main_genre'].value_counts().nlargest(15).index
movies_df_filtered_genres = movies_df[movies_df['main_genre'].isin(top_genres)]

# Create the boxplot
plt.figure(figsize=(15, 8))
sns.boxplot(x='main_genre', y='duration_minutes', data=movies_df_filtered_genres, palette='viridis')
plt.title('Distribución de Duración de Películas por Género Principal')
plt.xlabel('Género Principal')
plt.ylabel('Duración (minutos)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\nAnálisis de la Duración de Películas por Género:\n" \
      "El boxplot muestra la distribución de la duración de las películas para los géneros principales. \n" \
      "Podemos observar que géneros como 'Dramas' o 'Documentaries' tienden a tener una duración media más alta, \n" \
      "mientras que 'Children & Family Movies' suelen ser más cortas. \n" \
      "La variabilidad dentro de un mismo género también es notable, indicada por el tamaño de las cajas y la extensión de los bigotes. \n" \
      "Por ejemplo, 'Documentaries' y 'Dramas' a menudo muestran una mayor dispersión en sus duraciones, \n" \
      "sugiriendo una amplia gama de formatos dentro de estos géneros.")
```

```python
# Split the 'country' column to handle multiple countries per title and count occurrences
# Use .str.split(', ', expand=True) to split by comma and space, then stack to get individual country entries
countries_df = df['country'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True)

# Count the occurrences of each country and get the top 10
country_counts = countries_df.value_counts().nlargest(10)

# Plotting the top 10 countries with a bar chart
plt.figure(figsize=(12, 7))
sns.barplot(x=country_counts.index, y=country_counts.values, palette='plasma')
plt.title('Top 10 Países con Más Producciones en Netflix')
plt.xlabel('País')
plt.ylabel('Cantidad de Producciones')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()

print("\nAnálisis de los Países con Más Producciones:\n" \
      "El gráfico de barras muestra claramente que Estados Unidos domina el catálogo de Netflix \n" \
      "en términos de producciones, con una cantidad significativamente mayor que cualquier otro país. \n" \
      "Otros países como India y el Reino Unido también tienen una presencia importante. \n" \
      "Esta distribución es en gran medida esperada. No es sorprendente ver a Estados Unidos a la cabeza, \n" \
      "dada su posición como uno de los mayores productores de contenido audiovisual a nivel mundial \n" \
      "y el origen de Netflix. La fuerte presencia de India también es esperada, considerando su vasta \n" \
      "industria cinematográfica (Bollywood y otras). La distribución refleja la estrategia de Netflix \n" \
      "de ofrecer contenido global, pero con un claro énfasis en producciones de mercados clave.")
```

```python
import numpy as np

# Split the 'listed_in' column to handle multiple genres per title
genres_df = df.assign(listed_in=df['listed_in'].str.split(', ')).explode('listed_in')

# Clean up genre names (remove leading/trailing spaces)
genres_df['listed_in'] = genres_df['listed_in'].str.strip()

# Group by main_genre and type, then count titles
genre_type_counts = genres_df.groupby(['listed_in', 'type']).size().unstack(fill_value=0)

# Get the top N genres for better visualization (e.g., top 15 genres by total count)
top_genres = genre_type_counts.sum(axis=1).nlargest(15).index
genre_type_counts_filtered = genre_type_counts.loc[top_genres]

# Plotting the stacked bar chart
plt.figure(figsize=(14, 8))
genre_type_counts_filtered.plot(kind='bar', stacked=True, cmap='viridis', ax=plt.gca())
plt.title('Cantidad de Títulos por Género y Tipo (Película vs. Serie de TV)')
plt.xlabel('Género')
plt.ylabel('Cantidad de Títulos')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tipo')
plt.tight_layout()
plt.show()

print("\nAnálisis de Películas vs. Series por Género:\n" \
      "El gráfico de barras apiladas revela qué géneros están más asociados a películas o a series de TV.\n" \
      "Por ejemplo, géneros como 'Dramas', 'Comedies' y 'Documentaries' muestran una fuerte presencia en ambos tipos de contenido, \n" \
      "aunque con una inclinación hacia las películas en la mayoría de los casos. \n" \
      "En contraste, 'International TV Shows' y 'TV Dramas' son, como era de esperar, predominantemente series.\n" \
      "Mientras que 'Children & Family Movies' y 'Action & Adventure' tienden a ser más películas.\n" \
      "Esto sugiere que Netflix diversifica su catálogo de películas en una amplia gama de géneros, \n" \
      "pero concentra sus series en categorías específicas, muchas veces con un componente 'internacional' o 'dramático'.")
```

```python
import numpy as np

# Filter for movies only (this DataFrame should already be prepared from previous steps)
# Ensure 'duration_minutes' column is available
movies_df_scatter = df[df['type'] == 'Movie'].copy()

# Extract duration in minutes and convert to numeric, handling potential errors
movies_df_scatter['duration_minutes'] = movies_df_scatter['duration'].str.replace(' min', '', regex=False)
movies_df_scatter['duration_minutes'] = pd.to_numeric(movies_df_scatter['duration_minutes'], errors='coerce')
movies_df_scatter.dropna(subset=['duration_minutes'], inplace=True)

# Ensure release_year is numeric
movies_df_scatter['release_year'] = pd.to_numeric(movies_df_scatter['release_year'], errors='coerce')
movies_df_scatter.dropna(subset=['release_year'], inplace=True)

# Create the scatter plot
plt.figure(figsize=(14, 7))
sns.scatterplot(x='release_year', y='duration_minutes', data=movies_df_scatter, alpha=0.6)
plt.title('Relación entre Duración de Películas y Año de Lanzamiento')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Duración (minutos)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("\nAnálisis de la Relación entre Duración y Año de Lanzamiento:\n" \
      "El gráfico de dispersión muestra la distribución de la duración de las películas a lo largo de los años de lanzamiento.\n" \
      "A primera vista, no parece haber un cambio drástico en los estándares de duración promedio de las películas con el tiempo, \n" \
      "manteniéndose la mayoría entre 80 y 120 minutos, especialmente en las últimas décadas. \n" \
      "Sin embargo, se observa una mayor concentración de películas en los años más recientes, lo que es consistente con el crecimiento del catálogo de Netflix.\n" \
      "En cuanto a los outliers, se pueden identificar algunas películas con duraciones significativamente más largas \n" \
      "(más allá de los 200 minutos) o más cortas (menos de 40 minutos), distribuidas en diferentes años. \n" \
      "Estos podrían representar documentales extendidos, películas de arte o producciones experimentales que no siguen la duración estándar.\n" \
      "La presencia de un amplio rango de duraciones sugiere que Netflix ofrece variedad, \n" \
      "aunque el grueso del contenido se mantiene dentro de un rango convencional.")
```

## ⭐ Desafío Final

* Encuentra las combinaciones más frecuentes de **género + rating** y represéntalas en un **heatmap o gráfico de burbujas**.
* Pregunta guía: ¿Qué géneros están más dirigidos a un público adulto? ¿Y cuáles a público familiar o infantil?

```python
import numpy as np

# Prepare the data by splitting 'listed_in' and combining with 'rating'
genre_rating_df = df.assign(listed_in=df['listed_in'].str.split(', ')).explode('listed_in')

# Clean up genre names (remove leading/trailing spaces)
genre_rating_df['listed_in'] = genre_rating_df['listed_in'].str.strip()

# Handle potential missing ratings or genres by dropping NA rows
genre_rating_df.dropna(subset=['listed_in', 'rating'], inplace=True)

# Count occurrences of each genre-rating combination
combo_counts = genre_rating_df.groupby(['listed_in', 'rating']).size().unstack(fill_value=0)

# Filter for the most frequent genres (e.g., top 20 by total count) and ratings (to avoid clutter)
top_genres_for_heatmap = combo_counts.sum(axis=1).nlargest(20).index
top_ratings_for_heatmap = combo_counts.sum(axis=0).nlargest(10).index

# Filter the combo_counts DataFrame for selected genres and ratings
filtered_combo_counts = combo_counts.loc[top_genres_for_heatmap, top_ratings_for_heatmap]

# Plotting the heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(filtered_combo_counts, cmap='viridis', annot=True, fmt='d', linewidths=.5, linecolor='black')
plt.title('Frecuencia de Combinaciones de Género y Clasificación por Edad')
plt.xlabel('Clasificación por Edad')
plt.ylabel('Género')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

print("\nAnálisis de Combinaciones de Género y Clasificación por Edad:\n" \
      "El mapa de calor muestra las combinaciones más frecuentes de géneros y clasificaciones por edad. \n" \
      "**Géneros dirigidos a público adulto:** Se puede observar que géneros como 'Dramas', 'International Movies', \n" \
      "'Action & Adventure' y 'Comedies' tienen una alta frecuencia con clasificaciones como 'TV-MA' y 'TV-14', \n" \
      "lo que indica que gran parte de su contenido está dirigido a audiencias maduras o adolescentes mayores.\n" \
      "'Crime TV Shows' y 'Thrillers' también se asocian fuertemente con estas clasificaciones.\n" \
      "**Géneros dirigidos a público familiar o infantil:** Por otro lado, géneros como 'Kids' TV', \n" \
      "'Children & Family Movies' y 'Anime Series' muestran una clara predominancia en clasificaciones como 'TV-Y', 'TV-Y7', 'TV-G' y 'PG', \n" \
      "señalando su enfoque en audiencias más jóvenes. Esto refleja la segmentación de contenido de Netflix \n" \
      "para satisfacer las necesidades de diferentes grupos demográficos.")
```
